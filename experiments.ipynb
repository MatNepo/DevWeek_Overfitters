# YAKE
Как работает
YAKE (Yet Another Keyword Extractor) — это статистическая модель для извлечения ключевых слов из текста. Она анализирует текст без предварительного обучения, используя комбинацию статистических признаков: частоту слов, их позицию в тексте, длину предложений и контекст (соседние слова). Для каждого слова или фразы YAKE вычисляет "вес", где меньший вес означает более важное слово. Например, слова, которые часто встречаются в начале текста или в уникальном контексте, получают более высокий приоритет.

## Сильные стороны
Быстрая и лёгкая: не требует GPU или больших вычислительных ресурсов, подходит для работы на CPU в Colab.
Простая в использовании: не нужно обучать модель, достаточно задать язык и параметры.
Хорошо работает для коротких текстов.
Поддерживает разные языки: можно настроить для русского (lan="ru") или английского (lan="en").\
Гибкие параметры: можно регулировать длину фраз (n), разнообразие (dedupLim) и число хэштегов (top).
## Слабые стороны
Длинные хэштеги: при n=3 фразы могут быть громоздкими (например, #частьвидеопоказывает), что неудобно для соцсетей.
Ограниченная точность для смешанных текстов: при lan="ru" английские тексты (Text 5, Text 7) дают нерелевантные хэштеги (например, #The).
Статистический подход: не учитывает семантику, поэтому может пропустить контекстно важные слова. разговорные тексты или тексты с нестандартными словами могут давать менее релевантные результаты.
## Общая информация
YAKE разработана для автоматического извлечения ключевых слов в задачах обработки текстов, таких как аннотирование, поиск и классификация. Она особенно полезна, когда нужно быстро выделить основные темы без сложной настройки. В вашем случае YAKE подходит для генерации хэштегов, но требует настройки (например, n=2, dedupLim=0.7) для более лаконичных и релевантных результатов. Проблема с длинными хэштегами решается уменьшением n, а для смешанных текстов — использованием langdetect для выбора языка.

# spaCy с pytextrank
## Как работает
spaCy — это библиотека для обработки естественного языка, которая выполняет токенизацию, разметку частей речи (POS-тэгирование) и анализ текста.  Pytextrank — это дополнение к spaCy, реализующее алгоритм TextRank, основанный на графе. TextRank строит граф, где узлы — это слова или фразы, а рёбра — их связи (например, соседство в тексте). Алгоритм ранжирует узлы по важности, отдавая приоритет фразам, которые часто связаны с другими важными словами. В функции generate_hashtags_spacy текст обрабатывается spaCy, pytextrank извлекает ключевые фразы , и они преобразуются в хэштеги (например, #Монако).

## Сильные стороны
Учитывает структуру текста: благодаря POS-тэгированию и TextRank модель выделяет семантически важные фразы.
Гибкость: можно настроить части речи (pos_kept) для включения глаголов, что полезно для текстов про действия (Text 2, Text 13).
Поддерживает разные языки: смена модели (например, en_core_web_sm для английского) позволяет работать с Text 5 и Text 7.
Качество для структурированных текстов: хорошо работает для описаний с чёткими темами (Text 3 про концерт Blackpink).
## Слабые стороны
Ограничения ru_core_news_sm: маленькая модель плохо справляется с разговорными  или сложными текстами, что приводит к пустым результатам.
Медленнее YAKE: обработка графа TextRank и токенизация требуют больше времени
Пустые результаты: если текст не содержит фраз с высоким рангом, хэштеги не генерируются .

## Общая информация
spaCy широко используется для задач NLP, таких как извлечение сущностей, анализ зависимостей и классификация текста. Pytextrank добавляет возможности извлечения ключевых фраз, что делает его подходящим для генерации хэштегов.

# RAKE
## Как работает
RAKE (Rapid Automatic Keyword Extraction) — это алгоритм для извлечения ключевых фраз, основанный на анализе текста без обучения. RAKE разбивает текст на предложения (с помощью NLTK punkt_tab), выделяет слова, исключает стоп-слова (например, из NLTK stopwords) и формирует фразы, разделённые стоп-словами или знаками препинания. Каждой фразе присваивается вес на основе частоты слов и их связей (например, слова, часто встречающиеся вместе, получают больший вес).

## Сильные стороны
Простота и скорость: RAKE не требует обучения и работает быстрее, чем spaCy или KeyBERT.
Эффективность для чётких текстов: хорошо выделяет фразы в описаниях с явными темами .
Настраиваемые стоп-слова: добавление пользовательских стоп-слов (например, "это", "как") улучшает релевантность.
Лёгкий: минимальное потребление памяти, подходит для Colab без GPU.
## Слабые стороны
Зависимость от стоп-слов: неполный список стоп-слов приводит к нерелевантным хэштегам.
Ограниченная семантика: как и YAKE, RAKE не учитывает глубокий контекст, только частоту и связи слов.
Смешанные тексты: плохо справляется с английскими текстами без настройки языка (language="russian").
## Общая информация
RAKE разработан для быстрого извлечения ключевых фраз в задачах обработки текстов, таких как аннотирование и поиск. Он популярен благодаря простоте и скорости, но требует тщательной настройки стоп-слов и токенизатора и добавлению пользовательских стоп-слов. Установка max_length=2 сделает хэштеги более подходящими для соцсетей.

# KeyBERT
## Как работает
KeyBERT использует трансформерные модели (например, distiluse-base-multilingual-cased) для создания эмбеддингов текста и извлечения ключевых фраз. Эмбеддинги — это числовые представления слов или фраз, учитывающие их семантическое значение. KeyBERT сравнивает эмбеддинги текста с эмбеддингами кандидатных фраз (1–3 слова, keyphrase_ngram_range=(1, 3)) и выбирает те, которые ближе всего к смыслу текста (по косинусному сходству). Дополнительно может использоваться MMR (Maximal Marginal Relevance) для увеличения разнообразия фраз. В вашем коде KeyBERT извлекает до 10 фраз (top_n=10), которые преобразуются в хэштеги (например, #гитара ). Пустые результаты в вашем коде, вероятно, связаны с ошибкой инициализации модели или нехваткой ресурсов.

## Сильные стороны
Семантическая точность: благодаря трансформерам KeyBERT учитывает смысл текста, а не только статистику.
Поддержка смешанных текстов: модель distiluse или mpnet-base-v2 работает с русским и английским .
Гибкость: параметры like use_mmr и diversity позволяют контролировать разнообразие хэштегов.
Высокое качество для сложных текстов: подходит для описаний с неявными темами.
## Слабые стороны
Медленная и ресурсоёмкая: требует GPU для больших моделей (mpnet-base-v2), иначе может быть очень медленной.
Пустые результаты: ошибки инициализации или нехватка памяти (как в вашем случае) приводят к пустым спискам.
Чувствительность к настройкам: без use_mmr или правильного keyphrase_ngram_range фразы могут быть длинными или повторяющимися.
Сложность настройки: требует понимания параметров (например, diversity, stop_words).
## Общая информация
KeyBERT — это современный инструмент для извлечения ключевых слов, использующий мощь трансформеров. Он подходит для задач, где важна семантика, например, генерация хэштегов для социальных сетей или аннотирование контента. В вашем случае KeyBERT имеет потенциал для лучших результатов, но требует GPU и настройки (use_mmr=True, keyphrase_ngram_range=(1, 2)). Проблема с пустыми результатами решается использованием более мощной модели (paraphrase-multilingual-mpnet-base-v2) и обработкой ошибок.

```python
from langdetect import detect

def generate_hashtags_yake(texts, top_n=5):
    process = psutil.Process(os.getpid())
    mem_before = process.memory_info().rss / 1024 / 1024
    start_time = time.time()
    hashtags_list = []
    kw_extractor_ru = yake.KeywordExtractor(lan="ru", n=2, dedupLim=0.7, top=top_n)
    kw_extractor_en = yake.KeywordExtractor(lan="en", n=2, dedupLim=0.7, top=top_n)
    for text in texts:
        lang = detect(text)
        extractor = kw_extractor_en if lang == "en" else kw_extractor_ru
        keywords = extractor.extract_keywords(text)
        hashtags = ["#" + kw[0].replace(" ", "") for kw in keywords if len(kw[0]) <= 15]
        hashtags_list.append(hashtags)
    end_time = time.time()
    mem_after = process.memory_info().rss / 1024 / 1024
    total_time = end_time - start_time
    memory_used = mem_after - mem_before
    return hashtags_list, total_time, memory_used
```

```python
def generate_hashtags_spacy(texts, top_n=10, min_rank=0.05, max_words=2):
    process = psutil.Process(os.getpid())
    mem_before = process.memory_info().rss / 1024 / 1024
    start_time = time.time()
    hashtags_list = []
    for text in texts:
        # Очистка текста перед обработкой
        cleaned_text = scrubber(text)
        doc = nlp(cleaned_text)
        # Фильтрация фраз: учитываем top_n, min_rank, max_words и stop_words
        phrases = [
            p for p in doc._.phrases[:top_n]
            if p.rank > min_rank and len(p.text.split()) <= max_words
            and not any(word.lower() in stop_words for word in p.text.split())
            and len(p.text) <= 15
        ]
        hashtags = []
        for p in phrases:
            # Лемматизация для более универсальных хэштегов
            lemma = " ".join(nlp(word)[0].lemma_ for word in p.text.split())
            hashtag = "#" + "".join(word.capitalize() for word in lemma.split())
            hashtags.append(hashtag)
        hashtags_list.append(hashtags)
    end_time = time.time()
    mem_after = process.memory_info().rss / 1024 / 1024
    total_time = end_time - start_time
    memory_used = mem_after - mem_before
    return hashtags_list, total_time, memory_used
```

```python
def generate_hashtags_rake(texts, top_n=8):
    process = psutil.Process(os.getpid())
    mem_before = process.memory_info().rss / 1024 / 1024
    start_time = time.time()
    hashtags_list = []
    for text in texts:
        rake.extract_keywords_from_text(text)
        ranked_phrases = [p for p in rake.get_ranked_phrases()[:top_n] if len(p) <= 15]
        hashtags = ["#" + p.replace(" ", "") for p in ranked_phrases]
        hashtags_list.append(hashtags)
    end_time = time.time()
    mem_after = process.memory_info().rss / 1024 / 1024
    total_time = end_time - start_time
    memory_used = mem_after - mem_before
    return hashtags_list, total_time, memory_used
```

```python
def generate_hashtags_keybert(texts, top_n=8):
    process = psutil.Process(os.getpid())
    mem_before = process.memory_info().rss / 1024 / 1024
    start_time = time.time()
    hashtags_list = []
    custom_stop_words = stopwords.words('russian') + ['это', 'как', 'the', 'and']
    for text in texts:
        try:
            keywords = kw_model.extract_keywords(
                text,
                keyphrase_ngram_range=(1, 2),
                stop_words=custom_stop_words,
                top_n=top_n,
                use_mmr=True,
                diversity=0.7
            )
            hashtags = ["#" + kw[0].replace(" ", "") for kw in keywords if len(kw[0]) <= 15]
            hashtags_list.append(hashtags)
        except Exception as e:
            print(f"Error processing text with KeyBERT: {e}")
            hashtags_list.append([])
    end_time = time.time()
    mem_after = process.memory_info().rss / 1024 / 1024
    total_time = end_time - start_time
    memory_used = mem_after - mem_before
    return hashtags_list, total_time, memory_used
```
